#**Conditional (Şartlı) Pandas Operations**



```
turkey_series
```

Pekala, yukarıda biz bu değerleri bin üzerinden tanımlamıştık fakat bunların milyon olduğunu yorum satırında belirtmiştik, hemen sayısal olarak milyon hesabıan geçişi sağlayalım.

```
turkey_series * 1_000_000
```

```
turkey_series.mean()
```

Örnek olarak 5 milyondan fazla nüfuslu şehirleri boolean ifade olarak yazdıralım ekrana.

```
turkey_series > 5
```

Şimdi ise bu aldığımız beş şehrin ortalamasından büyük olanları ekrana yansıtalım.

```
turkey_series[turkey_series > turkey_series.mean()]
```

Standart Sapmamızı kontrol eddelim.

```
turkey_series.std()
```

3 milyondan fazla 8 milyondan az nüfuslu şehirleri listeleyelim.

```
turkey_series[(turkey_series > 3) & (turkey_series < 8)]
```

# PANDAS - DATA FRAME!

Şimdi pandas'ın en önemli kısımlarından biri olan Data frame konusuna göz atacağız. EN çok burayı kullanıp uygulayacağız. İlk başta kullandığımız Türkiye'nin en kalabalık 5 ili işimizi görecektir.
